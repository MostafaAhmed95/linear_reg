```python
import pandas as pd  
import numpy as np  
import matplotlib.pyplot as plt  
import seaborn as seabornInstance 
from sklearn.model_selection import train_test_split 
from sklearn.linear_model import LinearRegression
from sklearn import metrics
import math
%matplotlib inline
```

```python
data=open('case5.txt')

```

```python
x=[]
y=[]
for i in data:
    #print (i)
    point=i.rstrip().split(',')
    x.append(float(point[0]))
    y.append(float(point[1]))
```

```python
plt.scatter(x,y,s=1)  
plt.title('time vs acc reading')  
plt.xlabel('time')  
plt.ylabel('acc_reading')  
plt.show()
```

```python
x=np.array(x)
y=np.array(y)
```

```python
x=x.reshape(-1,1)
y=y.reshape(-1,1)
```

```python
regry=LinearRegression()
regry.fit(x,y)

```

```python
print(regry.intercept_)
print(regry.coef_)
```

```python
y_pred=regry.predict(x)
```

```python
plt.scatter(x, y,  color='gray')
plt.plot(x, y_pred, color='red', linewidth=2)
plt.show()
```

```python
#excluding the outliers
x=[]
y=[]
data=open('case5.txt')
for i in data:
    point=i.rstrip().split(',')
    if (float(point[1])>3900):
        x.append(float(point[0]))
        y.append(float(point[1]))
```

```python
x=np.array(x)
y=np.array(y)
x=x.reshape(-1,1)
y=y.reshape(-1,1)
```

```python
#fitting the data again
regry=LinearRegression()
regry.fit(x,y)
```

```python
#new intercet and coffecient
print(regry.intercept_)
print(regry.coef_)
```

```python
y_pred=regry.predict(x)
plt.scatter(x, y,  color='gray')
plt.plot(x, y_pred, color='red', linewidth=2)
plt.show()
```

```python
#i don't know what is meant by true value in this case 
# so i consider it the mean
mean = np.mean(y) #mean
std = np.std(y) #standard deviation
st_error = std / math.sqrt(y.size) #standard error
# For 99.9% confidence level from the Z tables z= 3.291
z = 3.291
ME = z * st_error #Margin error
print("my interval: ", mean+ME, " , ", mean-ME)

```
